```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(1,10,200)
```

```python
y=np.sin(x)
```

```python
# how to draw normal graph
plt.plot(x,y)
plt.xlabel("This is my X label")
plt.ylabel("This is my Y label")
plt.title("This is Sin wave")
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,"--r")
plt.grid()
```

```python
# how to draw scatter graph
x=np.random.rand(50)
y=np.random.rand(50)
color=np.random.rand(50)
size=1000*np.random.rand(50)
plt.scatter(x,y,c=color,s=size,alpha=.6,marker="v")
plt.xlabel("This is my X label")
plt.ylabel("This is my Y label")
plt.title("This is Random Data")
```

```python
#  how to draw bar graph
x=['x','y','z','a','b']
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel("This is my X label")
plt.ylabel("This is my Y label")
plt.title("This is Bar Data")
```

```python
# how to draw horizontal bar graph
plt.barh(x,y)
```

```python
# histogram graph
data=[1,2,3,4,5,1,2,2,5,8]
plt.hist(data)
```

```python
x=[1,2,6,4,5]
y=[6,2,8,6,0]
color=np.random.rand(5)
plt.scatter(x,y,c=color)
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z)
```

# Why is the subplot() function used?

```python
# The subplot() function takes three arguments that describes the layout of the figure.
# The layout is organized in rows and columns, which are represented by the first and second argument.
# The third argument represents the index of the current plot.
```

```python

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```
